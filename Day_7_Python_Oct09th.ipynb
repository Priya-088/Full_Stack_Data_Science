### Upsc, IAS, Goverment exams - Repetitive attempts

while i do not cross the cutoff I will keep on giving my exams

while condition is True:
    print()

````python
total_marks = 1000
cutoff = 400
scores = [100, 200, 300, 399, 500]

#I will keep in attempting untill I succeed

year = 0 #indexing starts from zero

while scores[year]< cutoff:
    print(f' Your score is " {scores[year]}, cutoff: {cutoff}" ')
    print('I will attempt next year')
    year = year+1 #Incrementing year to get in to current year
````

````python
#Train journey

stations = ['Station 01','Station 02','Station 03','Station 04']

current_station = 0
destination_station = 'Station 03'

while stations[current_station]!=destination_station:
    print(f'current station: {stations[current_station]}')
    print(f'my destination is: {destination_station}')
    print('continue the journey')
    current_station = current_station+1
    print(f'Next station is: {stations[current_station]}')
````

### break

````python
cost_of_items = [100, 200, 129, 456,1000,199,299]

total_sum = 0
budget = 2000

for cost in cost_of_items:
    total_sum = total_sum+cost
    print(f'current total: {total_sum}')
    if total_sum > budget:
        print(f'sunce current total is more than my budget: {budget}')
        total_sum = total_sum-cost
        break
        
        
print(f'Total amount to be paid: {total_sum}')
````

````python

for i in range(1,100):
    if i>50:
        break
    print(i)    
        
````

````python
for i in range(1,20):
    if i%2!=0:  
        continue  #start the loop again when the number is odd
    print(i)
````

````python
stations = ['Station 01','Station 02','Station 03','Station 04']

for current_station in stations:
    print(f'current station is: {current_station}')
    if current_station =='Station 02':
        break
````

````python
stations = ['Station 01','Station 02','Station 03','Station 04']

for current_station in stations:
    
    if current_station =='Station 02':
        continue
    print(f'current station is: {current_station}')
````

- Break: It completly stops the loop
- continue: It skips the current loop, it skips all the satement after 'continue' keywords
- pass: used when you dont want to write any statements currenlty and use it for future

````python
for i in range(1,10):
    pass  #if we want write code later we can use pass
````
