```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Problem Statement

A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
#importing the data

df_train = pd.read_csv('Datasets/train.csv')
df_train.head()
```

```python
# Importing the test data

df_test = pd.read_csv('Datasets/test.csv')
df_test.head()
```

```python
##append both train and test data

df = df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Since user ID is unique, we need to drop the column

df.drop(['User_ID'], axis=1, inplace=True)
```

```python
df
```

### Converting categorical to numerical

```python
#we can use one hot encoding using get_dummies
pd.get_dummies(df['Gender'])

#df['Gender']=pd.get_dummies(df['Gender'])
```

```python
df.head(5)
```

```python
#we can use map method to replace the values in 'Gender' column

df['Gender'] = df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
##we can use pd.get_dummies, but we end up creating more columns

pd.get_dummies(df['Age'],drop_first=True)
```

```python
#Handling categorical variable 'age'

df['Age'].unique()
```

```python
#We can use Target - Odinal encoding for the column 'age', where we can give priority
## to the age groups who shop more on eComm platform

df['Age'] = df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

#### Never use a '0' for odinal encoding, the mathematical equations coeffients may effect

```python
#Method - 2
#we can use 'preprocessing' library from sklearn
# from sklearn import preprocessing

# label_encoder = preprocessing.LabelEncoder()

# df['Age'] = label_encoder.fit_transform(df['Age'])
# df['Age']
```

```python
df.head(5)
```

```python
df['City_Category'].unique()
```

```python
#Handling 'City_Category' column
#if we use pd.get_dummies - then we can select n-1 colums for n labels in a particular column


df_city = pd.get_dummies(df['City_Category'], drop_first = True)

```

```python
df_city
```

```python
#need to combine df with df_city

df = pd.concat([df,df_city],axis=1)
df.head()
```

```python
#dropping 'City' categorical feature
df.drop(['City_Category'], axis=1, inplace=True)
```

```python
#replacing '4+' with '4'

df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.info()
```

```python
#converting 'Stay_In_Current_City_Years' to integer column

df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype('int')
```

```python
df.info()
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```
