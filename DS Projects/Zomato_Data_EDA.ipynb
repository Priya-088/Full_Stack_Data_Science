```python
df_country = pd.read_excel('Datasets/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
## to check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
#Maximum no.of trasactions happes in India
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
#which county has maximum transactions for zomato

plt.pie(country_val, labels=country_names)
```

```python
#Since the above chart is clumsy
#Top 3 countries where zomato is operated

plt.pie(country_val[:3],labels=country_names[:3])
```

```python
#to display percentage
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:
- Zomato maximum transactions are from India then it is followed by USA and then UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

##Observations
1. when rating between 4.5-4.9 --> Excellent
2. when rating between 4.0-4.5 --> Very good
3. When rating between 3.5-3.9 --> good
4. When rating between 2.5-3.4 --> average
5. When rating between 2.0-2.4 --> poor
6. 0 rating is given by many people

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating count',data=ratings,hue='Rating color')
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating count',data=ratings,hue='Rating color'
            ,palette=['blue','red','orange','yellow','green','green'])
```

Observations:
1. Not rated count is very high.
2. There is a gaussian distribution when we plot ratings againt rating count
3. maximum number of ratings are between 2.5 anf 3.4

Note: If there are missing ratings in the data, we can impute it with the average value between 2.5 and 3.4

```python
##count plot
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
